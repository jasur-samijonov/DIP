```python
import cv2
import matplotlib.pyplot as plt
image = cv2.imread('/Users/220712/Downloads/image.jpg')
image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
subtracted_image = cv2.subtract(image_RGB, 100)
plt.subplot(1, 2, 1)
plt.imshow(image_RGB)
plt.title('Original Image')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(subtracted_image)
plt.title('Subtracted Image (-100)')
plt.axis('off')
plt.show()
```

```python
import os
print(os.getcwd())
# I used it to get the location of the image(it was in the same folder with current working directory)
```

```python
R, G, B = cv2.split(image_RGB)
R_sub = cv2.subtract(R, 100)
red_reduced = cv2.merge((R_sub, G, B))
plt.subplot(1, 2, 1)
plt.imshow(image_RGB)
plt.title('Original Image')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(red_reduced)
plt.title('Red Channel Reduced (-100)')
plt.axis('off')
plt.show()
```

```python


added_image = cv2.add(image_RGB, 30)
plt.subplot(1, 2, 1)
plt.imshow(image_RGB)
plt.title('Original Image')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(added_image)
plt.title('Added Image (30)')
plt.axis('off')
plt.show()
```

```python

R_add = cv2.add(R, 100)
Red_bright = cv2.merge((R_add, G, B))
plt.subplot(1, 2, 1)
plt.imshow(image_RGB)
plt.title('Original Image')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(Red_bright)
plt.title('Red Bright Image (100)')
plt.axis('off')
plt.show()
```

```python
multiplied_image = cv2.multiply(image_RGB, 0.5)
plt.subplot(1,2,1)
plt.imshow(image_RGB)
plt.title('Original Image')
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(multiplied_image)
plt.title('Multiplied Image 0.5')
plt.axis('off')
plt.show()
```

```python
divided_image = cv2.divide(image_RGB, 0.5)
plt.subplot(1,2,1)
plt.imshow(image_RGB)
plt.title('Original Image')
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(divided_image)
plt.title('Divided Image 0.5')
plt.axis('off')
plt.show()
```
